### Exercise 10:
Write a Python function that takes two strings and checks if they are anagrams.

```python
# Exercise 10
```

````python
def teste_anagramme(chaine_1, chaine_2):
    if len(chaine_1) != len(chaine_2):
        return 0 # chaine_1 et chaone_2 ne sont pas des anagrammes.
    else:
        a = 0
        for i in range(len(chaine_1)):
            if chaine_1.count(chaine_1[i]) != chaine_2.count(chaine_1[i]):
                return 0 # chaine_1 et chaone_2 ne sont pas des anagrammes.
            else:
                a += 1
        if a == len(chaine_1):
            return 1 # chaine_1 et chaone_2 sont des anagrammes.

###############################################

def teste_anagramme_final(chaine_1, chaine_2):
    if teste_anagramme(chaine_1, chaine_2) == 1:
        print(f"\"{chaine_1}\" et \"{chaine_2}\" sont des anagrammes.")
    else:
        print(f"\"{chaine_1}\" et \"{chaine_2}\" ne sont pas des anagrammes.")

        
# Premier teste de la fonction ....

chaine_1 = "programmation"
chaine_2 = "groprammation"
teste_anagramme_final(chaine_1, chaine_2)

print("\n==========================================\n")

# Deuxième teste de la fonction ....

chaine_1 = "AWA"
chaine_2 = "AAW"
teste_anagramme_final(chaine_1, chaine_2)


print("\n==========================================\n")

# Troisiéme teste de la fonction ....

chaine_1 = "AWA"
chaine_2 = "AbW"
teste_anagramme_final(chaine_1, chaine_2)
````

### Exercise 11:
Write a Python program that checks if a number is prime.

```python
# Exercise 11
```

````python
def is_prime_0(nombre):
    a = 0
    for i in range(2, nombre//2):
        if nombre % i == 0:
            a += 1
    if a != 0:
        return 0 # pour dire que le nombre n'est pas premier
    else:
        return 1 # pour dire que le nombre est premier

##################################

def is_prime_1(nombre):
    if is_prime_0(nombre) == 1:
        print(f"{nombre} est un nombre premier.")
    else:
        print(f"{nombre} n'est pas un nombre premier.")


# Teste de la fonction...

nombre = 32
is_prime_1(nombre)
````

### Exercise 16:
Write a Python function that checks if a list contains any duplicates.

```python
# Exercise 16
```

````python
def compte_meme_element_dans_liste(liste, caractere):
    a = 0
    for i in liste:
        if caractere == i:
            a += 1
    return a

#######################################

def verif_doublons_liste(liste):
    i = 0
    j = 0
    while i < len(liste) and j == 0:
        if compte_meme_element_dans_liste(liste, liste[i]) > 1:
            print(f"Cette liste ({liste}) contient des doublons.")
            j = 1
        i += 1 
    if j == 0:
        print(f"Cette liste ({liste}) ne contient pas des doublons.")

# Teste 

liste = [12, 4, 5, 7]
verif_doublons_liste(liste)
````
